Notebook: Feature Selection and Data Cleaning
Purpose:
- Select relevant features for predicting energy_per_capita.
- Filter dataset by year and countries with sufficient data.
- Drop rows with missing target.
- Impute remaining missing values in GDP and energy share features.

```python
# -------------------------------
# 1. Imports
# -------------------------------
import pandas as pd

# -------------------------------
# 2. Load Dataset
# -------------------------------
url = "https://raw.githubusercontent.com/JiaLiiin/CS506-Project/refs/heads/proposal-update/Data/owid-energy-data-updated.csv"
df = pd.read_csv(url)

print(df.shape)
print(list(df.columns))
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
# -------------------------------
# 3. Feature Selection
# -------------------------------
selected_columns = [
    "country", "year",
    "population", "gdp",
    "energy_per_capita",
    "coal_cons_per_capita", "gas_energy_per_capita", "oil_energy_per_capita",
    "fossil_energy_per_capita", "fossil_share_energy",
    "renewables_energy_per_capita", "low_carbon_energy_per_capita",
    "renewables_share_energy", "low_carbon_share_energy",
    "hydro_share_energy", "solar_share_energy", "wind_share_energy"
]

df_selected = df[selected_columns].copy()

# Compute GDP per capita
df_selected["gdp_per_capita"] = df_selected["gdp"] / df_selected["population"]

df_selected.head()
```

```python
df_selected.groupby("year").apply(lambda x: x.isna().mean().mean())
```

```python
df_selected.groupby("country").apply(lambda x: x.isna().mean().mean()).sort_values(ascending=False)
```

```python
unique_countries = df_selected["country"].unique()
print(unique_countries)
```

```python
# -------------------------------
# 4. Filter by Year and Country
# -------------------------------
# Keep years >= 1965 (reliable coverage)
df_selected = df_selected[df_selected["year"] >= 1965]

# Drop countries with >40% missing
threshold = 0.4
country_missing = df_selected.groupby("country").apply(lambda x: x.isna().mean().mean())
good_countries = country_missing[country_missing <= threshold].index
df_selected = df_selected[df_selected["country"].isin(good_countries)]

# Drop aggregate entries
df_selected = df_selected.drop(df_selected[df_selected['country'].isin(['World', 'North America'])].index)
```

```python
# -------------------------------
# 5. Drop rows with missing target
# -------------------------------
df_selected = df_selected.dropna(subset=['energy_per_capita']).copy()
```

```python
df_selected.isna().mean().sort_values(ascending=False)
```

```python
# -------------------------------
# 6. Missing Value Imputation
# -------------------------------
# Sort by country/year first
df_selected = df_selected.sort_values(['country', 'year'])

# Interpolate GDP features
gdp_cols = ["gdp", "gdp_per_capita"]
df_selected[gdp_cols] = df_selected.groupby("country")[gdp_cols].transform(
    lambda x: x.interpolate().ffill().bfill()
)

# Forward/backward fill energy share features
share_cols = [
    "renewables_share_energy", "hydro_share_energy", "wind_share_energy",
    "fossil_share_energy", "low_carbon_share_energy", "solar_share_energy"
]
df_selected[share_cols] = df_selected.groupby("country")[share_cols].transform(
    lambda x: x.ffill().bfill()
)
```

```python
# -------------------------------
# 7. Final Check
# -------------------------------
print("Missing values per column after imputation:")
print(df_selected.isna().sum())

print("\nDataset shape:", df_selected.shape)
print("Number of countries:", df_selected['country'].nunique())
```
